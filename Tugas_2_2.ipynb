```python
customers = pd.read_csv('customers.csv')
gasstations = pd.read_csv('gasstations.csv')
products = pd.read_csv('products.csv')
yearmonth = pd.read_csv('yearmonth.csv')
transactions_1k = pd.read_csv('transactions_1k.csv')
display(customers, gasstations, products, yearmonth, transactions_1k)
```

```python
df = pd.merge(transactions_1k, customers)
df.rename(columns={'Segment':'Segment_x'},inplace=True)
df
```

```python
df1 = pd.merge(df, gasstations)
df1.rename(columns={'Segment':'Segment_y'},inplace=True)
df1
```

```python
df2 = pd.merge(df1, products)
df2
```

```python
yearmonth.rename(columns={'Date':'Date_z'},inplace=True)
df3 = pd.merge(df2,yearmonth)
df3
```
